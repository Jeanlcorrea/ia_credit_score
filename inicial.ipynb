```python
import pandas as pd

table = pd.read_csv("clientes.csv")

display(table)
display(table.info())
```

```python
from sklearn.preprocessing import LabelEncoder

codificador = LabelEncoder()

table["profissao"] = codificador.fit_transform(table["profissao"])

table["mix_credito"] = codificador.fit_transform(table["mix_credito"])

table["comportamento_pagamento"] = codificador.fit_transform(table["comportamento_pagamento"])
```

```python
display(table)
display(table.info())
```

```python
y = table["score_credito"]
x = table.drop(columns=["score_credito", "id_cliente"])

from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y)

from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

```python
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

from sklearn.metrics import accuracy_score

print(accuracy_score(y_teste, previsao_arvoredecisao))
print(accuracy_score(y_teste, previsao_knn))
```

```python
tabela_nova = pd.read_csv("novos_clientes.csv")

tabela_nova["profissao"] = codificador.fit_transform(tabela_nova["profissao"])

tabela_nova["mix_credito"] = codificador.fit_transform(tabela_nova["mix_credito"])

tabela_nova["comportamento_pagamento"] = codificador.fit_transform(tabela_nova["comportamento_pagamento"])

previsoes = modelo_arvoredecisao.predict(tabela_nova)

display(previsoes)
```

```python
tabela_II = pd.read_csv("outros_clientes.csv")

tabela_II["profissao"] = codificador.fit_transform(tabela_II["profissao"])

tabela_II["mix_credito"] = codificador.fit_transform(tabela_II["mix_credito"])

tabela_II["comportamento_pagamento"] = codificador.fit_transform(tabela_II["comportamento_pagamento"])

previsoes = modelo_arvoredecisao.predict(tabela_II)

display(tabela_II)
display(previsoes)
```

```python
tabela_III = pd.read_csv("tabela_III.csv")

tabela_III["profissao"] = codificador.fit_transform(tabela_III["profissao"])

tabela_III["mix_credito"] = codificador.fit_transform(tabela_III["mix_credito"])

tabela_III["comportamento_pagamento"] = codificador.fit_transform(tabela_III["comportamento_pagamento"])

previsoes = modelo_arvoredecisao.predict(tabela_III)

display(tabela_III)

previsoes_series = pd.Series(previsoes, name='Previsões')

# Criando um DataFrame com a tabela original e as previsões
resultado_final = pd.concat([tabela_III, previsoes_series], axis=1)

# Exibindo a tabela e as previsões de forma mais bonita
print(resultado_final.to_string(index=False))
```
